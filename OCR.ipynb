```python
%%writefile main.py
import cv2
import numpy as np
import torch
import io
from fastapi import FastAPI, UploadFile, File
from paddleocr import PaddleOCR
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image
from pdf2image import convert_from_bytes

app = FastAPI()

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"


processor = TrOCRProcessor.from_pretrained("microsoft/trocr-large-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-large-handwritten").to(DEVICE)

paddle_detector = PaddleOCR(
    use_angle_cls=True,
    lang='en',
    det=True,
    rec=False,
    show_log=False,
    det_db_unclip_ratio=2.0,
    det_db_thresh=0.3,
    det_db_box_thresh=0.5
)



def refine_character_visibility(crop):
    """Enhances image contrast and sharpness for better OCR accuracy."""
    gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
    denoised = cv2.bilateralFilter(gray, 9, 75, 75)
    gaussian_3 = cv2.GaussianBlur(denoised, (0, 0), 2.0)
    unsharp_image = cv2.addWeighted(denoised, 1.5, gaussian_3, -0.5, 0)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    enhanced = clahe.apply(unsharp_image)
    return cv2.cvtColor(enhanced, cv2.COLOR_GRAY2RGB)

def get_padded_crop(image, box, padding=15):
    """Crops the detected text area with extra padding."""
    points = np.array(box).astype(np.int32)
    x_min, y_min = np.min(points, axis=0)
    x_max, y_max = np.max(points, axis=0)
    h, w = image.shape[:2]
    y1, y2 = max(0, y_min - padding), min(h, y_max + padding)
    x1, x2 = max(0, x_min - padding), min(w, x_max + padding)
    return image[y1:y2, x1:x2]

def sort_boxes_into_lines(boxes, threshold=25):
    """Groups detection boxes into logical text lines based on Y-coordinate overlap."""
    if not boxes: return []

    boxes.sort(key=lambda b: b[0][1])

    lines = []
    current_line = [boxes[0]]
    for i in range(1, len(boxes)):
        if abs(boxes[i][0][1] - current_line[-1][0][1]) < threshold:
            current_line.append(boxes[i])
        else:

            current_line.sort(key=lambda b: b[0][0])
            lines.append(current_line)
            current_line = [boxes[i]]

    current_line.sort(key=lambda b: b[0][0])
    lines.append(current_line)
    return lines

def process_image(cv2_img):
    """Detects text regions and performs recognition on each."""
    detect_results = paddle_detector.ocr(cv2_img, cls=True)
    if not detect_results or not detect_results[0]:
        return []

    raw_boxes = [line[0] for line in detect_results[0]]
    line_groups = sort_boxes_into_lines(raw_boxes)

    processed_lines = []
    for line in line_groups:
        line_fragments = []
        for box in line:
            raw_crop = get_padded_crop(cv2_img, box)
            if raw_crop.size == 0: continue

            refined_crop = refine_character_visibility(raw_crop)
            pil_img = Image.fromarray(refined_crop)


            pixel_values = processor(images=pil_img, return_tensors="pt").pixel_values.to(DEVICE)
            with torch.no_grad():
                generated_ids = model.generate(
                    pixel_values,
                    max_length=128,
                    num_beams=10,
                    repetition_penalty=1.2
                )
                text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

            if text.strip():
                line_fragments.append(text.strip())

        if line_fragments:
            processed_lines.append(" ".join(line_fragments))

    return processed_lines


@app.post("/ocr")
async def extract_text(file: UploadFile = File(...)):
    contents = await file.read()
    all_text_lines = []

    if file.content_type == "application/pdf" or file.filename.lower().endswith(".pdf"):
        pages = convert_from_bytes(contents, dpi=300)
        for page in pages:

            open_cv_image = cv2.cvtColor(np.array(page), cv2.COLOR_RGB2BGR)
            page_lines = process_image(open_cv_image)

            all_text_lines.extend(page_lines)

    else:
        nparr = np.frombuffer(contents, np.uint8)
        image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
        image_lines = process_image(image)
        all_text_lines.extend(image_lines)

    return {
        "filename": file.filename,
        "total_lines": len(all_text_lines),
        "combined_output": all_text_lines
    }
```
